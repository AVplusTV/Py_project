Функция для чтения файла в таблицу с учетом возможности работы на разных платформах и заодно исследование данных в таблицах:


```python
def read_csv(link):
    link1 = 'datasets/' + link
    link2 = '/datasets/' + link
    try:
        name_link = pd.read_csv(link1)
    except:
        name_link = pd.read_csv(link2)
    display(name_link.head()) #просмотрим сразу первые пять строк
    display(name_link.info()) #просмотрим информацию    
    display(name_link.nunique()) # посчитаем уникальные значения
    display(name_link.columns) #проверим имена столбцов
    display('Количество явных дубликатов:', name_link.duplicated().sum()) #проверка на явные дубликаты
    display(name_link.describe()) #проверим численные данные
    display(name_link.corr()) #проверим корреляции
    return name_link
```

## Предобработка данных по результатам анализа

### Удаление явных дубликатов

```python
df = df.drop_duplicates().reset_index(drop = True)
```

```python
df.duplicated().sum()
```

### Удалим столбцы не влияющие на предсказания стоимости автомобиля

```python
df = df.drop(['NumberOfPictures', 'DateCrawled', 'DateCreated', 'LastSeen'], axis=1)
```

```python
df.head()
```

### Переименование столбцов

Так как требуется переименование всех столбцов

```python
df.set_axis(['price', 'vehicle_type', 'registration_year', 'gearbox',
       'power', 'model', 'kilometer', 'registration_month', 'fuel_type', 'brand',
       'not_repaired', 'postal_code'],axis = 'columns',inplace = True)
```

```python
df.columns
```

### Заполнение пустых значений в ячейках

#### Столбцы gearbox, model, postal_code

Эти данные не откуда брать и нечем особо заполнять, поэтому заполним все пропуски unknown

```python
df.head()
```

```python
df.info()
```

```python
df = df.fillna('unknown')
```

#### Столбец not_repaired

Здесь скорее всего владельцы не указали этот пункт потому что авто не ремонтировалось. Вместо unknow заполним no

```python
df['not_repaired'] = df['not_repaired'].replace('unknown', 'no')
```

#### Столбец vehicle_type

Есть отдельная категория other

```python
df['vehicle_type'] = df['vehicle_type'].replace('unknown', 'other')
```

#### Столбец fuel_type

Есть отдельная категория other

```python
df['fuel_type'] = df['fuel_type'].replace('unknown', 'other')
```

```python
display(df)
```

### Работа с максимальными и минимальными значениями числовых столбцов

#### Столбец price

В столбце цены есть автомобили с ценой 0, даже если допустить, что это сознательно владельцы отдают авто бесплатно, это никак не поможет в предсказании цены другого автомобиля, ну или скорее всего это неверные данные, и заполнять эти данные, как вариан можно при помощи модели линейной регрессии, но для обучения они нам будут только мешать. Посмотрим на них

```python
display(df[df['price'] == 0])
```

Удаляем их из таблицы

```python
df = df[df['price'] > 0]
display(df[df['price'] == 0])
```

Максимальное значение цены особо не чем не ограничено, но у нас в данных вполне разумная максимальная цена.

#### Столбец registration_year

Про максимальное значение года, логично предположить, что оно не может быть более текущего 2022 года и менее 1930

```python
display(df[df['registration_year'] > 2022])
```

```python
display(df[df['registration_year'] < 1930])
```

Удаляем

```python
df = df[df['registration_year'] < 2022]
df = df[df['registration_year'] > 1930]
```

здесь уже серьезное количество, и удалить мы их просто так не можем, нужно будет найти медианные значения по модели и заполнить эти данные

Напишем функцию для применения построчно

```python
median  = df.groupby('model')['power'].median().to_dict()
def fill_with_median (row):
    if row['power'] == 0:
        return median[row['model']]
    return row ['power']
df['power'] = df.apply(lambda row: fill_with_median(row), axis = 1)
```

###  О выдаче результатов в сводную таблицу

Для того чтобы выдать результат RMSE и TIME в сводную таблицу, придется два раза запускать обучение и подстчет ошибки - 1 раз для определения времени в переменную и второй раз для определения RMSE в переменную, по причине того, что метод %%timeit не выдает во вне ячейки никакие результаты кроме вывода времени.

### LinearRegression

```python
model_lr = LinearRegression(n_jobs=-1)# инициализируем модель LinearRegression
model_lr.fit(x_train, y_train) # обучим модель на тренировочной выборке
predictions_valid_lr = model_lr.predict(x_valid) # получим предсказания модели на валидационной выборке
result_lr = round(mse(y_valid, predictions_valid_lr)**0.5, 2) # RMSE
```

```python
%%timeit -n1 -r1 -o
model_lr = LinearRegression(n_jobs=-1)# инициализируем модель LinearRegression
model_lr.fit(x_train, y_train) # обучим модель на тренировочной выборке
predictions_valid_lr = model_lr.predict(x_valid) # получим предсказания модели на валидационной выборке
result_lr = mse(y_valid, predictions_valid_lr)**0.5 # RMSE
```

```python
result = _
time = str(result).split()
time_lr = time[0]+' '+time[1]
print('Время исполнения ячейки LinearRegression:', time_lr)
print('RMSE Модели LinearRegression:', result_lr)
```
