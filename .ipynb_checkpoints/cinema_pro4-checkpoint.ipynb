### Преобразования столбца director

То же самое проделаем и для режиссеров

```python
direct = list(df.groupby('director').groups.keys()) #список вариантов 
print(direct)
```

у первых 5-ти запесей уберем пробел спереди, далее у одного нет точки и лишние пробелы, лишние пробелы спереди и сзади нужно будет убирать в нескольких столбцах, поэтому создадим функцию:

```python
def del_spaces(column: str) -> None:
    n=0 #ограничим цикл при помощи условия, что уменьшит нагрузку
    list1 = list(df.groupby(column).groups.keys()) #список вариантов 
    for word in list1:
        if n < 20: # задаем 20 шагов цикла изменений (20 первых словосочетаний)
            del_space = word.strip() # удаляем пробелы в словосочетании
            df[column] = df[column].str.replace(word, del_space) # заменяем 
            n+=1
        else:
            break # прерываем цикл при достижении n шагов   
```

```python
# удалим пробелы в столбце 'director'
del_spaces('director')

    # можно убрать пробелы спереди и позади строки во всех ячейках, но быстрый просмотр не выявил еще таких строчек
df['director'] = df['director'].str.replace('а андрющенко', 'а.андрющенко')
df['director'] = df['director'].str.replace('а. калюжный', 'а.калюжный')
df['director'] = df['director'].str.replace('а. кутерье', 'а.кутерье')
```

Здесь есть одно важное замечание-предосторежение. Коэффициент корреляции Пирсона считается <b>только для непрерывных переменных</b>. К таким относятся стоимость, расстояние, площадь, средняя оценка рейтинга, бюдет, то есть все то, что может бють дробными. А, например, год является дискретной (прерывной) переменной, поэтому для него нужен уже другой статистический метод. Что касается столбцов с номером прокатного удостоверения или возрастным ограничением, то это и вовсе категориальные переменные, для которых коэффициент корреляции Пирсона не походит ни при каких условиях. Несмотря на то, что значения имеют целочисленный формат, по смыслу они являются названиями, а не числами.
    
Поэтому для подсчета и дальнейшей интерпретации коэффициента корреляции Пирсона нужно отбирать только непрерывные переменные.
</div>
