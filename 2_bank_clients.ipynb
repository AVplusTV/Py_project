```python
def median_total_income(list_income):
    for type in list_income:
        # –Ω–∞—Ö–æ–¥–∏–º –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –∏–≥–Ω–æ—Ä–∏—Ä—É—è –ø—Ä–æ–ø—É—Å–∫–∏
        median = df_income_check[df_income_check['income_type'] == type]['total_income'].median(skipna = True)
        # –∑–∞–º–µ–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏—è —è—á–µ–µ–∫ —Ñ–∏–ª—å—Ç—Ä—É—è –ø–æ —Ç–∏–ø—É –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ –∏ –Ω–∞–ª–∏—á–∏—é –ø—Ä–æ–ø—É—â–µ–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –¥–æ—Ö–æ–¥–∞—Ö
        df_income_check.loc[(df_income_check['income_type'] == type)
                            & (pd.isna(df_income_check['total_income'])),'total_income'] = median
        print(median)
```

–î–ª—è –∑–∞–º–µ–Ω—ã –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ purpose –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç–≤—É—é—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Å–ø–∏—Å–∫–∞ list_purpose = ['c–≤–∞–¥—å–±–∞', '–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å', '–∞–≤—Ç–æ–º–æ–±–∏–ª—å', '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'] –≤—ã–¥–µ–ª–∏–º —Å–ø–∏—Å–æ–∫ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ –∏ –ø—Ä–∏ –ø–æ–º–æ—â–∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏ —Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–π.

–ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–∏–æ—Ç–µ–∫—É –¥–ª—è –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏

```python
from pymystem3 import Mystem
m = Mystem()
```

–õ–µ–º–º–∞—Ç–∏–∑–∏—Ä—É–µ–º –∫–∞–∂–¥–æ–µ —É–Ω–∏–∫–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ list_unique_purpose –ò —Å–æ—Å—Ç–∞–≤–∏–º –Ω–æ–≤—ã–º —Å–æ–æ—Ç–≤–µ—Ç–≤–≤—É—é—â–∏–π—Å –ø–∏—Å–æ–∫ –∑–∞–º–µ–Ω  –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å–æ —Å–ø–∏—Å–∫–æ–º list_purpose = ['c–≤–∞–¥—å–±–∞', '–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å', '–∞–≤—Ç–æ–º–æ–±–∏–ª—å', '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ']

```python
list_lemmas = []
for idx in range(len(list_unique_purpose)):
    lemmas = m.lemmatize(list_unique_purpose[idx])
    if '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ' in lemmas:
        purpose = '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'
    elif '—Å–≤–∞–¥—å–±–∞' in lemmas:
        purpose = '—Å–≤–∞–¥—å–±–∞'
    elif '–∞–≤—Ç–æ–º–æ–±–∏–ª—å' in lemmas:
        purpose = '–∞–≤—Ç–æ–º–æ–±–∏–ª—å'
    else:
        purpose = '–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å'
    list_lemmas.append(purpose)
print(list_lemmas)
```

```python
list_debt_sum = [] #—Å–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–µ–º—â–∏–∫–æ–≤ —Å –ø—Ä–æ–±–ª–µ–º–∞–º–∏ –ø–æ –∫—Ä–µ–¥–∏—Ç—É –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
list_debt_count = [] #—Å–ø–∏—Å–æ–∫ –ø–æ–¥—Å—á–µ—Ç–∞ –≤—Å–µ–≥–æ –∑–∞–µ–º—â–∏–∫–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
list_debt_percent = []
#—Å–ø–∏—Å–æ–∫ –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –≤ –Ω–µ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–æ—Ü–µ–Ω—Ç–∞ –∑–∞–µ–º—â–∏–∫–æ–≤ —Å –ø—Ä–æ–±–ª–µ–º–∞–º–∏ –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
for quantity in list_children:
    debt_sum = df_children[df_children['children'] == quantity]['debt'].sum()
    debt_count = df_children[df_children['children'] == quantity]['debt'].count()
    debt_percent = round(debt_sum / debt_count * 100, 1)
    list_debt_sum.append(debt_sum)
    list_debt_count.append(debt_count)
    list_debt_percent.append(debt_percent)
print(list_debt_sum)  
print(list_debt_count) 
print(list_debt_percent)
```

–ü—Ä–∏ –ø–æ–º–æ—â–∏ —Ü–∏–∫–ª–∞ –≤—ã–≤–µ–¥–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ü–µ–ª–µ–π –∫—Ä–µ–¥–∏—Ç–∞

```python
list_debt_sum = [] #—Å–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–µ–º—â–∏–∫–æ–≤ —Å –ø—Ä–æ–±–ª–µ–º–∞–º–∏ –ø–æ –∫—Ä–µ–¥–∏—Ç—É –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
list_debt_count = [] #—Å–ø–∏—Å–æ–∫ –ø–æ–¥—Å—á–µ—Ç–∞ –≤—Å–µ–≥–æ –∑–∞–µ–º—â–∏–∫–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
list_debt_percent = []
#—Å–ø–∏—Å–æ–∫ –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –≤ –Ω–µ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–æ—Ü–µ–Ω—Ç–∞ –∑–∞–µ–º—â–∏–∫–æ–≤ —Å –ø—Ä–æ–±–ª–µ–º–∞–º–∏ –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
for purpose in list_purpose:
    debt_sum = df_purpose[df_purpose['purpose'] == purpose]['debt'].sum()
    debt_count = df_purpose[df_purpose['purpose'] == purpose]['debt'].count()
    debt_percent = round(debt_sum / debt_count * 100, 1)
    list_debt_sum.append(debt_sum)
    list_debt_count.append(debt_count)
    list_debt_percent.append(debt_percent)
print(list_debt_sum)  
print(list_debt_count) 
print(list_debt_percent)
```
